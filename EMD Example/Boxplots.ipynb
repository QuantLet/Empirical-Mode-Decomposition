```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df = pd.read_excel("/BirdvsUAV.xlsx")
```

```python
df
```

```python
# mean instantaneous frequency boxplots
sns.set(style="whitegrid")
feature_pairs = [("m_inst_pos_1", "m_inst_neg_1"), ("m_inst_pos_2", "m_inst_neg_2"), ("m_inst_pos_3", "m_inst_neg_3")]
IMF_no = ["IMF 1", "IMF 2", "IMF 3"]

fig, axes = plt.subplots(1, 3, figsize=(18, 6), facecolor="none")

for ax in axes:
    ax.set_facecolor("none")
for i, (f_pos, f_neg) in enumerate(feature_pairs):
    # DataFrame fÃ¼r Plot reorganisieren
    df_long = pd.DataFrame({
        "Wert": pd.concat([df[f_pos], df[f_neg]], ignore_index=True),
        "Anteil": ["Positiv"] * len(df[f_pos]) + ["Negativ"] * len(df[f_neg]),
        "Label": ["Bird" if l == 0 else "UAV" for l in df["uav"]] * 2
    })

    # Boxplot mit Seaborn
    sns.boxplot(data=df_long, x="Label", y="Wert", hue="Anteil",
                palette=["magenta", "green"], ax=axes[i])
    axes[i].set_title(f"{IMF_no[i]}", fontsize = 18)
    axes[i].set_xlabel("")
    axes[i].set_ylabel("")
    axes[i].legend_.remove()

plt.tight_layout()
plt.show()
```
