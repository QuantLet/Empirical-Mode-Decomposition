```python
import numpy as np
import matplotlib.pyplot as plt

# Define radius
radius = 40

# Define angle theta
theta = np.radians(30) 

# calculate coordinates
x_point = radius * np.cos(theta)
y_point = radius * np.sin(theta)

# plot figure
plt.figure(figsize=(8, 8))

# plot coordinate system
plt.axhline(0, color='black',linewidth=1)
plt.axvline(0, color='black',linewidth=1)

# concentric circles
num_circles = 6 
circle_radii = np.linspace(10, 60, num_circles)

# add circles
for radius in circle_radii:
    circle = plt.Circle((0, 0), radius, color='b', fill=False, linestyle='--', linewidth=0.8)
    plt.gca().add_artist(circle)

# Plot target
plt.scatter(x_point, y_point, c='r', marker='o') 

# plot the line from the origin
plt.plot([0, x_point], [0, y_point], c='g')

# add axis labels
plt.xlabel("X")
plt.ylabel("Y")
plt.axis('equal')

# define range
plt.xlim(-60, 60)
plt.ylim(-60, 60)

# plot entire image
plt.savefig("/radar_polar.png",transparent = True)
plt.show()
```
